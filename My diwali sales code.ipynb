# Exploratory Data Analysis

## Gender

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data = df)
```

```python
#Categorising in Male and Female

ax = sns.countplot(x = 'Gender' , data = df)
for count in ax.containers:
    ax.bar_label(count)
```

```python
#Total money spent by each of the genders
df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False)
```

```python
sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Gender' , y = 'Amount' , data = sales_gen)
```

#### From above graph it is evident that most of the buyers are females and their purchasing power is relatively much higher

## Age

```python
df.columns
```

```python
#it will still run without hue but wont be this much descriptive
ay = sns.countplot( x = 'Age Group' , hue = 'Gender',data = df)
```

```python
ay = sns.countplot(data = df , x = 'Age Group' , hue = 'Gender')
for bars in ay.containers:
    ay.bar_label(bars)
```

```python
df.columns
```

```python
#Total Amount vs Age Group
sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False)
sns.barplot(x = 'Age Group' , y = 'Amount' , data = sales_age)
```

#### From here we can see that majority buyers are between age group 26-35 and they are females

### State

```python
df.columns
```

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.barplot(data = sales_state , x = 'State' , y = 'Orders')
sns.set(rc={'figure.figsize' :(28,8)})
```

```python
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.barplot(data = sales_state , x = 'State' , y = 'Amount')
sns.set(rc={'figure.figsize' : (17,7)})
```

* From above graph we can see from which states majority sales belong to Uttar Pradesh, Maharashtra , Karnataka

## Marital Status

```python
df.columns
```

```python
ax = sns.countplot(data = df , x = 'Martial_Status')
# Here we change the dimensions otherwise it will take (15,5) as earlier told
sns.set(rc = {'figure.figsize' : (5,5)})
for count in ax.containers:
    ax.bar_label(count)
```

```python
sales_state = df.groupby(['Martial_Status' , 'Gender'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False)

sns.set(rc = {'figure.figsize' : (6,5)})
sns.barplot(data = sales_state , x = 'Martial_Status' , y = 'Amount' , hue = 'Gender')
```

* From the above graph it is evident that the maximum purchasing power is possesed by the married women

## Occupation

```python
sns.set(rc ={'figure.figsize' : (20,5)})
ax = sns.countplot(data = df , x = 'Occupation')
for count in ax.containers:
    ax.bar_label(count)
```

```python
sales_state = df.groupby(['Occupation'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False)

sns.set(rc = {'figure.figsize' : (20,5)})
sns.barplot(data = sales_state , x = 'Occupation' , y = 'Amount')
```

* From the above graph we can see that most of the buyers are from IT, Healthcare and Aviation Sector

## Product Category

```python
df.columns
```

```python
sns.set(rc = {'figure.figsize' : (28,8)})
ax = sns.countplot(data = df , x = 'Product_Category')
for count in ax.containers:
    ax.bar_label(count)
```

```python
sales_state = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False).head(10)
sns.set(rc = {'figure.figsize' : (28,8)})
sns.barplot(data = sales_state , x = 'Product_Category', y = 'Amount') 
```

* Most products sold are from Food, Clothing and Electronics Industry

```python
sales_state = df.groupby(['Product_ID'] , as_index = False)['Orders'].sum().sort_values(by = 'Orders' , ascending = False).head(10)
sns.set(rc = {'figure.figsize' : (28,8)})
sns.barplot(data = sales_state , x = 'Product_ID' , y = 'Orders')
```

```python
# The above chart can also be prepared in an alternate way which would be
fig1 , ax1 = plt.subplots(figsize = (12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending = False).plot(kind = 'bar')
```
